```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
```

```python
ex = 'European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices'
```

```python
ne_tree = ne_chunk(pos_tag(word_tokenize(ex)))
print(ne_tree)
```

```python
def preprocess(text):
    sent = pos_tag(word_tokenize(ex))
    return sent
```

```python
sent = preprocess(ex)
sent
```

```python
pattern = 'NP: {<DT>?<JJ>*<NN>}'
```

```python
chunk_phrase = nltk.RegexpParser(pattern)
chunk_sent = chunk_phrase.parse(sent)
print(chunk_sent)
chunk_sent.draw()
```

```python
from nltk.chunk import conllstr2tree, tree2conlltags
from pprint import pprint
```

```python
iob_tagged = tree2conlltags(chunk_sent)
iob_tagged
```

```python
import spacy
from spacy import displacy
from collections import Counter
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
doc = nlp('European authorities fined Google a record $5.1 billion on Wednesday for abusing its power in the mobile phone market and ordered the company to alter its practices')
```

```python
print([(X.text, X.label_) for X in doc.ents])
```

```python
pprint([(X, X.ent_iob_, X.ent_type_) for X in doc])
```

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
```

```python
ny_bb = url_to_string('https://www.nytimes.com/2018/08/13/us/politics/peter-strzok-fired-fbi.html?hp&action=click&pgtype=Homepage&clickSource=story-heading&module=first-column-region&region=top-news&WT.nav=top-news')
article = nlp(ny_bb)
len(article.ents)
```

```python
labels = [X.label_ for X in article.ents]
Counter(labels)
```

```python
items = [X.text for X in article.ents]
Counter(items).most_common(3)
```

```python
sentences = [sent for sent in article.sents]
print(sentences[20])
```

```python
displacy.render(nlp(str(sentences[20])), style='ent', jupyter=True)
```

```python
displacy.render(nlp(str(sentences[20])), style='dep', jupyter=True, options={'distance': 60})
```

```python
pprint([(x.orth_, x.pos_, x.lemma_) for x in [y for y in nlp(str(sentences[20])) if not y.is_stop and y.pos_ != 'PUNCT']])
```
